```python
# Upload: StudentsPerformance.csv and gym_membership.csv

# Load data
students = pd.read_csv('/content/drive/MyDrive/colab dataset/StudentsPerformance.csv')

gym = pd.read_csv('/content/drive/MyDrive/colab dataset/gym_membership.csv')
gym.tail(5)
```

Bar Chart

```python
# Count gender
gender_count = students['gender'].value_counts()

# Simple bar chart
plt.bar(['Female', 'Male'], gender_count.values)
plt.title('Gender Count')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.show()
```

```python
# With color
plt.bar(['Female', 'Male'], gender_count.values, color=['red', 'green'])
plt.ylabel('Count')
plt.title('Students by Gender')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.show()
```

1. Bar chart of gym membership **types**

```python
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()
abonoment_type_counts = gym['abonoment_type'].value_counts()
plt.bar(abonoment_type_counts.index, abonoment_type_counts.values)
plt.title('Bar chart of gym membership types')
plt.xlabel('Gym membership types----->')
plt.ylabel('count of types----->')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.show()
```

**2. Histogram of time in gym**

```python
# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
plt.hist(gym['avg_time_in_gym'], bins=12, color='green', edgecolor='white')
plt.title('Histogram of time in gym')
plt.xlabel('Time in gym')
plt.ylabel('Frequency')
```

**3. Scatter plot: Age vs visits per week**

```python
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
plt.scatter(gym['Age'], gym['visit_per_week'])
plt.title('Age vs Visits per Week')
plt.xlabel('Age')
plt.ylabel('Visits per Week')
plt.show()
```

**1. Box plot: Gender vs Age**

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')
plt.figure(figsize=(8, 6))
sns.boxplot(data=gym, x='gender', y='Age')
plt.title('Gender vs Age')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

**2. Count plot: Days per week**

```python
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')

plt.figure(figsize=(8, 6))
sns.countplot(data=gym, x='days_per_week')
plt.title('Days per Week')
plt.xlabel('Days')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True,which='both',linestyle='-',linewidth=0.2,alpha=0.7)
plt.show()
```

**3. Bar plot: Average visits by membership type**

```python
# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.title('Average visits by membership type')
plt.xlabel('Membership Type')
plt.ylabel('Visits Time')
plt.tight_layout()
plt.grid(True,which='both',linestyle='-',linewidth=0.2,alpha=0.7)
plt.show()
```
